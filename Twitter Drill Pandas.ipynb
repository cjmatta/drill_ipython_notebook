#### Twitter Drill View
```
create or replace view mfs.views.twitter_tweets as select
CAST(t.`dir3` as INT) as `hour`,
CAST(t.`dir2` as INT) as `day`,
CAST(t.`dir1` as INT) as `month`,
CAST(t.`dir0` as INT) as `year`,
CAST(t.`id` as BIGINT) as `id`,
CAST(t.`user`.`id` as BIGINT) as `user_id`,
CAST(t.`text` as VARCHAR(140)) as `tweet`,
to_timestamp(t.`created_at`, 'EEE MMM dd HH:mm:ss Z YYYY') as `created_at`,
CAST(t.`favorited` as BOOLEAN) as `favorited`,
CAST(t.`favorite_count` as INT) as `favorite_count`,
CAST(t.`retweeted` as BOOLEAN) as `retweeted`,
CAST(t.`retweet_count` as INT) as `retweet_count`,
CAST(t.`possibly_sensitive` as BOOLEAN) as `possibly_sensetive`,
CAST(t.`place` as VARCHAR(140)) as `place`,
t.`coordinates`,
CAST(t.`source` as VARCHAR(140)) as `source`
from mfs.`/user/cmatta/tweets` t;
```

````python
# Looking at tweet volume of a search term
# Apple Tweets
apple_sql = """
    SELECT
    CONCAT(t.`year`, '-', t.`month`, '-', t.`day`),
    COUNT(t.`id`)
    FROM mfs.views.`twitter_tweets` t
    WHERE t.`tweet` LIKE '%apple%'
    GROUP BY t.`year`, t.`month`, t.`day`
    ORDER BY t.`year`, t.`month`, t.`day`
"""
cursor.execute(apple_sql)
apple_data = cursor.fetchall()

# Samsung Tweets
samsung_sql = """
    SELECT
    CONCAT(t.`year`, '-', t.`month`, '-', t.`day`),
    COUNT(t.`id`)
    FROM mfs.views.`twitter_tweets` t
    WHERE t.`tweet` LIKE '%samsung%'
    GROUP BY t.`year`, t.`month`, t.`day`
    ORDER BY t.`year`, t.`month`, t.`day`
"""
cursor.execute(samsung_sql)
samsung_data = cursor.fetchall()
````

#### Pandas
We will be creating a [DataFrame](http://pandas.pydata.org/pandas-docs/dev/generated/pandas.DataFrame.html) using [Pandas](http://pandas.pydata.org/)
> class `pandas.DataFrame(data=None, index=None, columns=None, dtype=None, copy=False)`
Two-dimensional size-mutable, potentially heterogeneous tabular data structure with labeled axes (rows and columns). Arithmetic operations align on both row and column labels. Can be thought of as a dict-like container for Series objects. The primary pandas data structure

````python
# tweets = []
# time_index = []
# for tweet in apple_data:
#     tweets.append(tweet[1])
#     time_index.append(datetime.datetime.strptime(tweet[0], "%Y-%m-%d"))

# tweet_volume = pandas.DataFrame(tweets, index=time_index, columns=["Apple"])
    
tweet_volume = pandas.DataFrame(
    [x[1] for x in apple_data],
    index=[datetime.datetime.strptime(x[0], "%Y-%m-%d") for x in apple_data],
    columns=["Apple"]
)
````

````python
tweet_volume['Samsung'] = pandas.Series([x[1] for x in samsung_data], index=tweet_volume.index)
````

````python
tweet_volume
````

````python
tweet_volume.plot(kind='bar', title="Apple v Samsung tweets per day")
````

````python
# iPhone Tweets
iphone_sql = """
    SELECT
    CONCAT(t.`year`, '-', t.`month`, '-', t.`day`, ' ', t.`hour`, ':00'),
    COUNT(t.`id`)
    FROM mfs.views.`twitter_tweets` t
    WHERE t.`tweet` LIKE '%iphone%'
    GROUP BY t.`year`, t.`month`, t.`day`, t.`hour`
    ORDER BY t.`year`, t.`month`, t.`day`, t.`hour`
"""
cursor.execute(iphone_sql)
iphone_data = cursor.fetchall()

# Android Tweets
android_sql = """
    SELECT
    CONCAT(t.`year`, '-', t.`month`, '-', t.`day`, ' ', t.`hour`, ':00'),
    COUNT(t.`id`)
    FROM mfs.views.`twitter_tweets` t
    WHERE t.`tweet` LIKE '%android%'
    GROUP BY t.`year`, t.`month`, t.`day`, t.`hour`
    ORDER BY t.`year`, t.`month`, t.`day`, t.`hour`
"""
cursor.execute(android_sql)
android_data = cursor.fetchall()
````

````python
mobileos_tweet_volume = pandas.DataFrame([x[1] for x in iphone_data], index=[datetime.datetime.strptime(x[0], "%Y-%m-%d %H:%M") for x in iphone_data], columns=["iPhone"])
````

````python
mobileos_tweet_volume["Android"] = pandas.Series([x[1] for x in android_data], index=[datetime.datetime.strptime(x[0], "%Y-%m-%d %H:%M") for x in android_data])
````

````python
mobileos_tweet_volume.plot(kind='bar', title="iPhone v. Android", figsize=[24, 8])
````
